```python
import pandas as pd

gamestop_csv = 'gamestop_data_2021_12.csv'
column_oder = [
    'title',
    'description',
    'sku',
    'category',
    'sub_cateory'
    'brand',
    'price',
    'images',
]

games = pd.read_csv(gamestop_csv, encoding='utf-8')

games = games[column_oder]
games
```

```python
games = games.rename(columns={
    'title': 'name',
    'brand': 'brand',
    'price': 'price',
    'sku': 'sku',
    'images': 'image_url',
    'description': 'description',
    'category': 'category',
    'sub_cateory': 'sub_category'
})
games
```

```python
from bs4 import BeautifulSoup

games['description'] = games['description'].apply(lambda x: BeautifulSoup(str(x), 'html.parser').get_text())
games
```

```python
games['image_url'] = games['image_url'].apply(lambda x: x.split('~')[0] if pd.notnull(x) else '')
games
```

```python

output_csv = 'gamestop_data_formatted.csv'
games.to_csv(output_csv, index=False)

games
```
